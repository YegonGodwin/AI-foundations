```python
def simple_chatbot():
    print("Chatbot: Hi! I'm a simple chatbot. Type 'bye' to exit.")
    
    while True:
        user_input = input("You: Hello").lower() 
        
        if "hello" in user_input or "hi" in user_input:
            print("Chatbot: Hello! How can I help you?")
        elif "how are you" in user_input:
            print("Chatbot: I'm just a program, but thanks for asking!")
        elif "bye" in user_input:
            print("Chatbot: Goodbye! Have a great day!")
            break
        else:
            print("Chatbot: I didn't understand that. Can you rephrase?")

# Run the chatbot
simple_chatbot()
```

```python
import tkinter as tk
from tkinter import scrolledtext, messagebox, ttk
import nltk
from nltk.chat.util import Chat, reflections
from tkinter import font as tkfont
import random

class ModernEduBot:
    def __init__(self):
        # Initialize NLTK
        try:
            nltk.data.find('tokenizers/punkt')
        except LookupError:
            nltk.download('punkt')

        # Modern color scheme
        self.colors = {
            'primary': '#4a6fa5',
            'secondary': '#166088',
            'accent': '#4fc3f7',
            'background': '#f8f9fa',
            'text': '#212529',
            'user_bubble': '#e3f2fd',
            'bot_bubble': '#ffffff',
            'user_text': '#0d47a1',
            'bot_text': '#263238'
        }

        # Enhanced conversation pairs
        self.conversation_pairs = [
            [r"hi|hello|hey", ["Hello! üëã Ready to learn something new today?", 
                              "Hi there! What would you like to explore?",
                              "Greetings! I'm here to help you learn."]],
            [r"how are you?", ["I'm always excited to teach! üí°", 
                             "Doing great - ready to share knowledge!",
                             "100% operational and ready to help!"]],
            [r"what is your name?", ["I'm EduBot, your AI learning companion! ü§ñ", 
                                    "You can call me EduBot!"]],
            [r"bye|goodbye", ["Goodbye! Come back soon to learn more! üëã", 
                            "See you later! Keep learning!",
                            "Have a wonderful day! üåü"]],

            # Learning domains with emojis
            [r"math|mathematics", ["üßÆ Math is fascinating! What specific topic? Algebra? Calculus?"]],
            [r"programming|coding", ["üíª Programming is powerful! Python? Java? JavaScript?"]],
            [r"science", ["üî¨ Science is amazing! Physics? Chemistry? Biology?"]],
            [r"history", ["üèõ History teaches us so much! Any particular era?"]],
            [r"web development", ["üåê Web dev is in demand! Frontend? Backend?"]],

            # Fallback
            [r"(.*)", ["I'm always learning - could you rephrase that? ü§î", 
                      "Interesting! Tell me more about that.",
                      "Let me think about that..."]]
        ]

        self.chatbot = Chat(self.conversation_pairs, reflections)

    def create_gui(self):
        """Create modern GUI with improved styling"""
        self.window = tk.Tk()
        self.window.title("EduBot - AI Learning Assistant")
        self.window.geometry("750x800")
        self.window.configure(bg=self.colors['background'])
        self.window.resizable(True, True)

        # Custom fonts
        self.title_font = tkfont.Font(family='Segoe UI', size=16, weight='bold')
        self.chat_font = tkfont.Font(family='Segoe UI', size=11)
        self.button_font = tkfont.Font(family='Segoe UI', size=10, weight='bold')

        # Header frame
        self._create_header()

        # Chat frame with scrollbar
        self._create_chat_frame()

        # Input frame
        self._create_input_frame()

        # Menu
        self._create_menu()

        # Welcome message
        self._display_welcome()

        return self.window

    def _create_header(self):
        """Create modern header with logo"""
        header = tk.Frame(self.window, bg=self.colors['primary'], height=80)
        header.pack(fill='x')

        # Logo and title
        logo_label = tk.Label(
            header,
            text="üß† EduBot",
            font=self.title_font,
            bg=self.colors['primary'],
            fg='white',
            padx=15,
            pady=15
        )
        logo_label.pack(side='left')

        # Subtitle
        subtitle = tk.Label(
            header,
            text="Your AI Learning Assistant",
            font=('Segoe UI', 10),
            bg=self.colors['primary'],
            fg='white'
        )
        subtitle.pack(side='left', padx=10)

    def _create_chat_frame(self):
        """Create modern chat display area"""
        chat_container = tk.Frame(self.window, bg=self.colors['background'])
        chat_container.pack(expand=True, fill='both', padx=20, pady=10)

        # Chat history with custom styling
        self.chat_history = scrolledtext.ScrolledText(
            chat_container,
            wrap=tk.WORD,
            width=80,
            height=25,
            font=self.chat_font,
            bg=self.colors['background'],
            padx=15,
            pady=15,
            relief=tk.FLAT,
            highlightthickness=0
        )
        self.chat_history.pack(expand=True, fill='both')
        self.chat_history.tag_config('user', 
                                   foreground=self.colors['user_text'],
                                   lmargin1=50, 
                                   lmargin2=50,
                                   rmargin=50)
        self.chat_history.tag_config('bot', 
                                   foreground=self.colors['bot_text'],
                                   lmargin1=50, 
                                   lmargin2=50,
                                   rmargin=50)
        self.chat_history.configure(state='disabled')

    def _create_input_frame(self):
        """Create modern input area with send button"""
        input_frame = tk.Frame(self.window, bg=self.colors['background'])
        input_frame.pack(fill='x', padx=20, pady=(0, 20))

        # Modern entry field
        self.user_input = ttk.Entry(
            input_frame,
            font=self.chat_font,
            style='Modern.TEntry'
        )
        self.user_input.pack(side='left', expand=True, fill='x', padx=(0, 10))
        self.user_input.bind('<Return>', self._send_message)

        # Styled send button
        style = ttk.Style()
        style.configure('Modern.TButton', 
                       font=self.button_font,
                       background=self.colors['primary'],
                       foreground='white',
                       borderwidth=0)

        send_button = ttk.Button(
            input_frame,
            text="Send",
            command=self._send_message,
            style='Modern.TButton'
        )
        send_button.pack(side='right')

    def _create_menu(self):
        """Create modern menu bar"""
        menubar = tk.Menu(self.window, bg=self.colors['background'], fg=self.colors['text'])
        self.window.config(menu=menubar)

        # Help menu with modern styling
        help_menu = tk.Menu(menubar, tearoff=0, 
                          bg=self.colors['background'],
                          fg=self.colors['text'],
                          activebackground=self.colors['accent'])
        menubar.add_cascade(label="Help", menu=help_menu)
        help_menu.add_command(label="About", command=self._show_about)
        help_menu.add_command(label="Clear Chat", command=self._clear_chat)
        help_menu.add_separator()
        help_menu.add_command(label="Exit", command=self.window.quit)

    def _display_welcome(self):
        """Display styled welcome message"""
        welcome_msg = (
            "EduBot: Welcome to your personalized learning experience! üåü\n\n"
            "I can help you with:\n"
            "‚Ä¢ Mathematics üßÆ\n"
            "‚Ä¢ Programming üíª\n"
            "‚Ä¢ Science üî¨\n"
            "‚Ä¢ History üèõ\n"
            "‚Ä¢ And much more!\n\n"
            "What would you like to learn today?"
        )
        
        self._add_message(welcome_msg, 'bot')

    def _send_message(self, event=None):
        """Process and display messages with modern bubbles"""
        user_message = self.user_input.get().strip()
        if not user_message:
            return

        # Add user message
        self._add_message(f"You: {user_message}", 'user')
        
        # Get and add bot response
        bot_response = self.chatbot.respond(user_message)
        if isinstance(bot_response, list):
            bot_response = random.choice(bot_response)
        self._add_message(f"EduBot: {bot_response}", 'bot')
        
        # Clear input
        self.user_input.delete(0, tk.END)

    def _add_message(self, message, sender):
        """Add styled message to chat"""
        self.chat_history.configure(state='normal')
        
        # Configure tags based on sender
        if sender == 'user':
            bg_color = self.colors['user_bubble']
            tag = 'user'
        else:
            bg_color = self.colors['bot_bubble']
            tag = 'bot'
        
        # Insert message with proper styling
        self.chat_history.insert(tk.END, message + "\n\n", tag)
        self.chat_history.configure(state='disabled')
        self.chat_history.see(tk.END)

    def _show_about(self):
        """Show modern about dialog"""
        about_text = (
            "EduBot - AI Learning Assistant\n\n"
            "Version 2.0\n"
            "Developed with Python and NLTK\n\n"
            "Features:\n"
            "‚Ä¢ Interactive learning\n"
            "‚Ä¢ Multiple subjects\n"
            "‚Ä¢ Modern interface\n\n"
            "¬© 2023 EduBot Project"
        )
        messagebox.showinfo("About EduBot", about_text)

    def _clear_chat(self):
        """Clear chat history"""
        self.chat_history.configure(state='normal')
        self.chat_history.delete(1.0, tk.END)
        self._display_welcome()
        self.chat_history.configure(state='disabled')

    def run(self):
        """Run application"""
        self.create_gui()
        self.window.mainloop()

if __name__ == "__main__":
    try:
        bot = ModernEduBot()
        bot.run()
    except Exception as e:
        messagebox.showerror("Error", f"An error occurred: {str(e)}")
```

```python
import tkinter as tk
from tkinter import scrolledtext
import nltk
from nltk.chat.util import Chat, reflections
import random
import colorsys

class ModernChatbotUI:
    def __init__(self):
        # Download NLTK resources
        try:
            nltk.data.find('tokenizers/punkt')
        except LookupError:
            nltk.download('punkt')

        # Color Palette
        self.colors = {
            'background': '#2C3E50',
            'primary': '#3498DB',
            'secondary': '#2ECC71',
            'text': '#ECF0F1',
            'accent': '#E74C3C'
        }

        # Expanded conversation pairs
        self.conversation_pairs = [
            [r"hi|hello|hey", [
                "üåü Hi there! Ready to explore knowledge?", 
                "üëã Greetings! What shall we learn today?", 
                "üöÄ Hello! I'm your intelligent learning companion!"
            ]],
            [r"how are you?", [
                "üí° Charged up and ready to help you learn!", 
                "üîã Always energetic and eager to assist!", 
                "üåà Feeling fantastic and full of knowledge!"
            ]],
            [r"(.*) help", [
                "ü§ñ I'm here to help! What topic interests you?", 
                "üìö Learning is my passion. How can I guide you?", 
                "üß† Ask away! Knowledge is just a message away!"
            ]],
            [r"bye|goodbye", ["Goodbye!", "Bye!"]],
            [r"(.*)", ["Let's explore that idea further!", "Interesting perspective!"]]
        ]

        self.chatbot = Chat(self.conversation_pairs, reflections)

    def _hsv_to_rgb(self, h, s, v):
        """Convert HSV to RGB values (0-255 scale)"""
        r, g, b = colorsys.hsv_to_rgb(h, s, v)
        return int(r * 255), int(g * 255), int(b * 255)

    def create_gradient_background(self, window):
        """Create dynamic gradient background"""
        canvas = tk.Canvas(window, width=window.winfo_width(), height=window.winfo_height())
        canvas.place(x=0, y=0, relwidth=1, relheight=1)
        
        hue = random.random()
        for i in range(100):
            r1, g1, b1 = self._hsv_to_rgb(hue, 0.6, 0.8)
            r2, g2, b2 = self._hsv_to_rgb(hue + 0.1, 0.6, 0.6)
            
            r = r1 + (r2 - r1) * i / 100
            g = g1 + (g2 - g1) * i / 100
            b = b1 + (b2 - b1) * i / 100
            
            color = f'#{int(r):02x}{int(g):02x}{int(b):02x}'
            canvas.create_rectangle(0, i, window.winfo_width(), i+1, fill=color, outline="")

    def create_fancy_gui(self):
        """Create modern GUI layout"""
        window = tk.Tk()
        window.title("ü§ñ KnowledgeBot")
        window.geometry("700x800")
        
        # Gradient background
        self.create_gradient_background(window)

        # Chat container
        chat_frame = tk.Frame(window, bg='#ffffff10')
        chat_frame.place(relwidth=0.9, relheight=0.8, relx=0.05, rely=0.1)

        # Chat history
        chat_history = scrolledtext.ScrolledText(
            chat_frame,
            wrap=tk.WORD,
            font=('Arial', 12),
            bg='#ffffff15',
            fg=self.colors['text'],
            insertbackground=self.colors['accent']
        )
        chat_history.pack(expand=True, fill='both')
        chat_history.configure(state='disabled')

        # Input field
        input_frame = tk.Frame(window, bg='#ffffff10')
        input_frame.place(relwidth=0.9, relheight=0.1, relx=0.05, rely=0.85)

        user_input = tk.Entry(
            input_frame,
            font=('Arial', 12),
            bg='#ffffff20',
            fg=self.colors['text']
        )
        user_input.pack(side=tk.LEFT, fill='both', expand=True, padx=5)

        # Send button
        send_button = tk.Button(
            input_frame,
            text="‚û§",
            font=('Arial', 14),
            command=lambda: self.send_message(user_input, chat_history),
            bg=self.colors['primary'],
            fg=self.colors['text'],
            activebackground=self.colors['secondary']
        )
        send_button.pack(side=tk.RIGHT)

        user_input.bind('<Return>', lambda event: self.send_message(user_input, chat_history))
        self.display_welcome_message(chat_history)
        
        return window

    def display_welcome_message(self, chat_history):
        """Show initial welcome message"""
        chat_history.configure(state='normal')
        chat_history.insert(tk.END, "üîÆ Welcome to KnowledgeBot!\n\n", 'bot')
        chat_history.tag_config('bot', foreground=self.colors['secondary'])
        chat_history.configure(state='disabled')

    def send_message(self, input_entry, chat_history):
        """Handle user input and responses"""
        user_message = input_entry.get().strip()
        if not user_message:
            return

        chat_history.configure(state='normal')
        chat_history.insert(tk.END, f"You: {user_message}\n", 'user')
        response = self.chatbot.respond(user_message) or "Let's explore that idea!"
        chat_history.insert(tk.END, f"Bot: {response}\n\n", 'bot')
        
        # Configure tags
        chat_history.tag_config('user', foreground=self.colors['primary'])
        chat_history.tag_config('bot', foreground=self.colors['secondary'])
        
        input_entry.delete(0, tk.END)
        chat_history.configure(state='disabled')
        chat_history.see(tk.END)

    def run(self):
        """Start the application"""
        window = self.create_fancy_gui()
        window.mainloop()

if __name__ == "__main__":
    chatbot = ModernChatbotUI()
    chatbot.run()
```

```python
%pip install numpy

import nltk
import random
import numpy as np # type: ignore
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
import json
import pickle
import re

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

class AdvancedChatbot:
    def __init__(self):
        # Lemmatizer for word normalization
        self.lemmatizer = WordNetLemmatizer()
        
        # Conversation patterns and intents
        self.intents = {
            'greetings': {
                'patterns': [
                    'hi', 'hello', 'hey', 'greetings', 'wassup', 
                    'good morning', 'good afternoon', 'good evening'
                ],
                'responses': [
                    'Hello! How can I assist you today?', 
                    'Hi there! What can I help you with?', 
                    'Greetings! I\'m ready to chat.'
                ]
            },
            'help': {
                'patterns': [
                    'help', 'assist me', 'what can you do', 
                    'support', 'guidance'
                ],
                'responses': [
                    'I can help you with various tasks. Ask me anything!',
                    'I\'m an AI assistant designed to provide information and support.',
                    'I\'m here to help. What do you need assistance with?'
                ]
            },
            'mood': {
                'patterns': [
                    'how are you', 'how\'s it going', 'what\'s up', 
                    'how do you feel'
                ],
                'responses': [
                    'I\'m doing great, thanks for asking!', 
                    'Feeling awesome and ready to help!', 
                    'Always good and eager to assist you!'
                ]
            },
            'unknown': {
                'patterns': [],
                'responses': [
                    'I\'m not sure I understand. Could you rephrase that?',
                    'That\'s an interesting query. Could you provide more context?',
                    'I\'m still learning. Can you explain that differently?'
                ]
            }
        }
        
        # Neural network model configuration
        self.max_words = 1000
        self.max_len = 20
        
    def preprocess_text(self, text):
        """
        Advanced text preprocessing
        """
        # Convert to lowercase
        text = text.lower()
        
        # Remove special characters
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        
        # Tokenize
        tokens = nltk.word_tokenize(text)
        
        # Lemmatization
        tokens = [self.lemmatizer.lemmatize(token) for token in tokens]
        
        return ' '.join(tokens)
    
    def create_training_data(self):
        """
        Prepare training data for the neural network
        """
        # Collect all patterns
        training_sentences = []
        training_labels = []
        
        for intent, data in self.intents.items():
            for pattern in data['patterns']:
                training_sentences.append(self.preprocess_text(pattern))
                training_labels.append(intent)
        
        # Tokenization
        tokenizer = Tokenizer(num_words=self.max_words, oov_token='<OOV>')
        tokenizer.fit_on_texts(training_sentences)
        
        # Convert text to sequences
        sequences = tokenizer.texts_to_sequences(training_sentences)
        padded_sequences = pad_sequences(sequences, maxlen=self.max_len, padding='post')
        
        # Encode labels
        label_encoder = LabelEncoder()
        encoded_labels = label_encoder.fit_transform(training_labels)
        
        return padded_sequences, encoded_labels, tokenizer, label_encoder
    
    def build_model(self, input_shape, num_classes):
        """
        Create neural network model
        """
        model = keras.Sequential([
            keras.layers.Embedding(self.max_words, 16, input_length=input_shape),
            keras.layers.GlobalAveragePooling1D(),
            keras.layers.Dense(24, activation='relu'),
            keras.layers.Dense(num_classes, activation='softmax')
        ])
        
        model.compile(
            optimizer='adam', 
            loss='sparse_categorical_crossentropy', 
            metrics=['accuracy']
        )
        
        return model
    
    def train_chatbot(self):
        """
        Train the neural network
        """
        # Prepare training data
        X, y, tokenizer, label_encoder = self.create_training_data()
        
        # Build and train model
        model = self.build_model(X.shape[1], len(np.unique(y)))
        model.fit(X, y, epochs=50, verbose=0)
        
        return model, tokenizer, label_encoder
    
    def get_response(self, user_input):
        """
        Generate chatbot response
        """
        # Preprocess input
        processed_input = self.preprocess_text(user_input)
        
        # Tokenize and pad input
        input_sequence = self.tokenizer.texts_to_sequences([processed_input])
        padded_input = pad_sequences(input_sequence, maxlen=self.max_len, padding='post')
        
        # Predict intent
        prediction = self.model.predict(padded_input)
        intent_index = np.argmax(prediction)
        intent = self.label_encoder.inverse_transform([intent_index])[0]
        
        # Select response
        if intent in self.intents:
            return random.choice(self.intents[intent]['responses'])
        else:
            return random.choice(self.intents['unknown']['responses'])
    
    def interactive_chat(self):
        """
        Interactive chat interface
        """
        print("ü§ñ Advanced AI Chatbot: Type 'bye' to exit")
        
        # Train model
        self.model, self.tokenizer, self.label_encoder = self.train_chatbot()
        
        while True:
            user_input = input("You: ")
            
            if user_input.lower() == 'bye':
                print("Chatbot: Goodbye! Have a great day!")
                break
            
            response = self.get_response(user_input)
            print("Chatbot:", response)

# Run the chatbot
if __name__ == "__main__":
    chatbot = AdvancedChatbot()
    chatbot.interactive_chat()
```
