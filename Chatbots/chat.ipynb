```python
def simple_chatbot():
    print("Chatbot: Hi! I'm a simple chatbot. Type 'bye' to exit.")
    
    while True:
        user_input = input("You: Hello").lower() 
        
        if "hello" in user_input or "hi" in user_input:
            print("Chatbot: Hello! How can I help you?")
        elif "how are you" in user_input:
            print("Chatbot: I'm just a program, but thanks for asking!")
        elif "bye" in user_input:
            print("Chatbot: Goodbye! Have a great day!")
            break
        else:
            print("Chatbot: I didn't understand that. Can you rephrase?")

# Run the chatbot
simple_chatbot()
```

```python
import nltk
from nltk.chat.util import Chat, reflections

# Define pairs of patterns and responses
pairs = [
    [r"hi|hello|hey", ["Hello!", "Hi there!"]],
    [r"how are you?", ["I'm just a bot, but I'm doing well!"]],
    [r"what is your name?", ["I'm a simple chatbot!"]],
    [r"bye|goodbye", ["Goodbye!", "Bye!"]],
    [r"(.*)", ["I didn't get that. Can you ask something else?"]]
]

# Create the chatbot
chatbot = Chat(pairs, reflections)

def run_chatbot():
    print("Chatbot: Hi! Type 'bye' to exit.")
    chatbot.converse()

# Run the chatbot
run_chatbot()
```

```python
import tkinter as tk
from tkinter import scrolledtext, messagebox
import nltk
from nltk.chat.util import Chat, reflections

class EducationalChatbot:
    def __init__(self):
        """
        Initialize the chatbot with predefined conversation pairs
        """
        # Download NLTK resources if not already downloaded
        try:
            nltk.data.find('tokenizers/punkt')
        except LookupError:
            nltk.download('punkt')

        # Comprehensive conversation pairs
        self.conversation_pairs = [
            # Greetings
            [r"hi|hello|hey", ["Hello!", "Hi there!", "Greetings!"]],
            [r"how are you?", ["I'm great!", "Doing well, thanks for asking!", "Always ready to help!"]],
            [r"what is your name?", ["I'm EduBot, your educational assistant!", "Call me EduBot!"]],
            [r"bye|goodbye", ["Goodbye!", "See you later!", "Have a great day!"]],

            # Learning Domains
            [r"Learn maths", ["Math is awesome! What specific math topic interests you?"]],
            [r"Learn programming languages", ["Programming is exciting! Which language would you like to explore?"]],
            [r"Learn science", ["Science is fascinating! What scientific field intrigues you?"]],
            [r"Learn history", ["History helps us understand the world! What period or topic fascinates you?"]],
            [r"Learn web development", ["Web development is a great skill! What aspect would you like to learn?"]],

            # Specific Topics
            [r"Addition", ["Addition is combining numbers to find their total sum. Want to learn more?"]],
            [r"Subtraction", ["Subtraction involves removing one number from another. Need more details?"]],
            
            # Programming Languages
            [r"Python", ["Python: A versatile language used in web development, data science, and AI!"]],
            [r"Java", ["Java: Powerful for enterprise and Android app development!"]],
            [r"JavaScript", ["JavaScript: The language of web interactivity!"]],

            # Science Domains
            [r"Physics", ["Physics explores the fundamental laws of the universe!"]],
            [r"Chemistry", ["Chemistry investigates matter and its interactions!"]],
            [r"Biology", ["Biology studies life in all its incredible forms!"]],

            # Fallback
            [r"(.*)", ["I'm still learning. Could you rephrase that?", "Interesting! Tell me more."]]
        ]

        # Initialize chat engine
        self.chatbot = Chat(self.conversation_pairs, reflections)

    def create_gui(self):
        """
        Create the graphical user interface for the chatbot
        """
        # Main window setup
        self.window = tk.Tk()
        self.window.title("EduBot: Educational Chatbot")
        self.window.geometry("600x700")
        self.window.configure(bg='#f0f0f0')

        # Chat history display
        self.chat_history = self._create_chat_history()

        # User input section
        self.user_input = self._create_user_input()

        # Send button
        self._create_send_button()

        # Menu bar
        self._create_menu_bar()

        # Welcome message
        self._display_welcome_message()

        return self.window

    def _create_chat_history(self):
        """
        Create scrollable chat history text area
        """
        chat_history = scrolledtext.ScrolledText(
            self.window, 
            wrap=tk.WORD, 
            width=70, 
            height=25, 
            font=('Arial', 10),
            bg='white',
            state='disabled'
        )
        chat_history.pack(padx=15, pady=10)
        return chat_history

    def _create_user_input(self):
        """
        Create user input entry field
        """
        input_frame = tk.Frame(self.window, bg='#f0f0f0')
        input_frame.pack(padx=15, pady=5, fill='x')

        user_input = tk.Entry(
            input_frame, 
            width=50, 
            font=('Arial', 12),
            bd=2,
            relief=tk.GROOVE
        )
        user_input.pack(side=tk.LEFT, expand=True, fill='x', padx=(0, 10))
        user_input.bind('<Return>', self._send_message)  # Enable Enter key
        return user_input

    def _create_send_button(self):
        """
        Create send message button
        """
        send_button = tk.Button(
            self.window, 
            text="Send", 
            command=self._send_message,
            bg='#4CAF50',
            fg='white',
            font=('Arial', 10, 'bold')
        )
        send_button.pack(pady=5)

    def _create_menu_bar(self):
        """
        Create menu bar with additional options
        """
        menubar = tk.Menu(self.window)
        self.window.config(menu=menubar)

        # Help menu
        help_menu = tk.Menu(menubar, tearoff=0)
        menubar.add_cascade(label="Help", menu=help_menu)
        help_menu.add_command(label="About", command=self._show_about)
        help_menu.add_command(label="Clear Chat", command=self._clear_chat)

    def _display_welcome_message(self):
        """
        Display initial welcome message
        """
        self.chat_history.configure(state='normal')
        welcome_msg = "EduBot: Welcome! I'm your educational assistant. What would you like to learn today?\n"
        self.chat_history.insert(tk.END, welcome_msg)
        self.chat_history.configure(state='disabled')

    def _send_message(self, event=None):
        """
        Process and send user message
        """
        user_message = self.user_input.get().strip()
        
        if not user_message:
            return

        # Update chat history
        self.chat_history.configure(state='normal')
        self.chat_history.insert(tk.END, f"You: {user_message}\n")
        
        # Get chatbot response
        bot_response = self.chatbot.respond(user_message)
        self.chat_history.insert(tk.END, f"EduBot: {bot_response}\n")
        
        # Clear input and scroll to bottom
        self.user_input.delete(0, tk.END)
        self.chat_history.configure(state='disabled')
        self.chat_history.see(tk.END)

    def _show_about(self):
        """
        Display about information
        """
        messagebox.showinfo(
            "About EduBot", 
            "EduBot: An Educational Chatbot\n\n"
            "Developed to help you learn various topics interactively!\n"
            "Version 1.0\n"
            "Created with Python and Tkinter"
        )

    def _clear_chat(self):
        """
        Clear chat history
        """
        self.chat_history.configure(state='normal')
        self.chat_history.delete(1.0, tk.END)
        self._display_welcome_message()
        self.chat_history.configure(state='disabled')

    def run(self):
        """
        Run the chatbot GUI
        """
        self.create_gui()
        self.window.mainloop()

def main():
    """
    Main function to run the chatbot
    """
    try:
        chatbot = EducationalChatbot()
        chatbot.run()
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
from tkinter import scrolledtext
import nltk
from nltk.chat.util import Chat, reflections
import random
import colorsys

class ModernChatbotUI:
    def __init__(self):
        # Download NLTK resources
        try:
            nltk.data.find('tokenizers/punkt')
        except LookupError:
            nltk.download('punkt')

        # Color Palette
        self.colors = {
            'background': '#2C3E50',
            'primary': '#3498DB',
            'secondary': '#2ECC71',
            'text': '#ECF0F1',
            'accent': '#E74C3C'
        }

        # Expanded conversation pairs
        self.conversation_pairs = [
            [r"hi|hello|hey", [
                "🌟 Hi there! Ready to explore knowledge?", 
                "👋 Greetings! What shall we learn today?", 
                "🚀 Hello! I'm your intelligent learning companion!"
            ]],
            [r"how are you?", [
                "💡 Charged up and ready to help you learn!", 
                "🔋 Always energetic and eager to assist!", 
                "🌈 Feeling fantastic and full of knowledge!"
            ]],
            [r"(.*) help", [
                "🤖 I'm here to help! What topic interests you?", 
                "📚 Learning is my passion. How can I guide you?", 
                "🧠 Ask away! Knowledge is just a message away!"
            ]],
            [r"bye|goodbye", ["Goodbye!", "Bye!"]],
            [r"(.*)", ["Let's explore that idea further!", "Interesting perspective!"]]
        ]

        self.chatbot = Chat(self.conversation_pairs, reflections)

    def _hsv_to_rgb(self, h, s, v):
        """Convert HSV to RGB values (0-255 scale)"""
        r, g, b = colorsys.hsv_to_rgb(h, s, v)
        return int(r * 255), int(g * 255), int(b * 255)

    def create_gradient_background(self, window):
        """Create dynamic gradient background"""
        canvas = tk.Canvas(window, width=window.winfo_width(), height=window.winfo_height())
        canvas.place(x=0, y=0, relwidth=1, relheight=1)
        
        hue = random.random()
        for i in range(100):
            r1, g1, b1 = self._hsv_to_rgb(hue, 0.6, 0.8)
            r2, g2, b2 = self._hsv_to_rgb(hue + 0.1, 0.6, 0.6)
            
            r = r1 + (r2 - r1) * i / 100
            g = g1 + (g2 - g1) * i / 100
            b = b1 + (b2 - b1) * i / 100
            
            color = f'#{int(r):02x}{int(g):02x}{int(b):02x}'
            canvas.create_rectangle(0, i, window.winfo_width(), i+1, fill=color, outline="")

    def create_fancy_gui(self):
        """Create modern GUI layout"""
        window = tk.Tk()
        window.title("🤖 KnowledgeBot")
        window.geometry("700x800")
        
        # Gradient background
        self.create_gradient_background(window)

        # Chat container
        chat_frame = tk.Frame(window, bg='#ffffff10')
        chat_frame.place(relwidth=0.9, relheight=0.8, relx=0.05, rely=0.1)

        # Chat history
        chat_history = scrolledtext.ScrolledText(
            chat_frame,
            wrap=tk.WORD,
            font=('Arial', 12),
            bg='#ffffff15',
            fg=self.colors['text'],
            insertbackground=self.colors['accent']
        )
        chat_history.pack(expand=True, fill='both')
        chat_history.configure(state='disabled')

        # Input field
        input_frame = tk.Frame(window, bg='#ffffff10')
        input_frame.place(relwidth=0.9, relheight=0.1, relx=0.05, rely=0.85)

        user_input = tk.Entry(
            input_frame,
            font=('Arial', 12),
            bg='#ffffff20',
            fg=self.colors['text']
        )
        user_input.pack(side=tk.LEFT, fill='both', expand=True, padx=5)

        # Send button
        send_button = tk.Button(
            input_frame,
            text="➤",
            font=('Arial', 14),
            command=lambda: self.send_message(user_input, chat_history),
            bg=self.colors['primary'],
            fg=self.colors['text'],
            activebackground=self.colors['secondary']
        )
        send_button.pack(side=tk.RIGHT)

        user_input.bind('<Return>', lambda event: self.send_message(user_input, chat_history))
        self.display_welcome_message(chat_history)
        
        return window

    def display_welcome_message(self, chat_history):
        """Show initial welcome message"""
        chat_history.configure(state='normal')
        chat_history.insert(tk.END, "🔮 Welcome to KnowledgeBot!\n\n", 'bot')
        chat_history.tag_config('bot', foreground=self.colors['secondary'])
        chat_history.configure(state='disabled')

    def send_message(self, input_entry, chat_history):
        """Handle user input and responses"""
        user_message = input_entry.get().strip()
        if not user_message:
            return

        chat_history.configure(state='normal')
        chat_history.insert(tk.END, f"You: {user_message}\n", 'user')
        response = self.chatbot.respond(user_message) or "Let's explore that idea!"
        chat_history.insert(tk.END, f"Bot: {response}\n\n", 'bot')
        
        # Configure tags
        chat_history.tag_config('user', foreground=self.colors['primary'])
        chat_history.tag_config('bot', foreground=self.colors['secondary'])
        
        input_entry.delete(0, tk.END)
        chat_history.configure(state='disabled')
        chat_history.see(tk.END)

    def run(self):
        """Start the application"""
        window = self.create_fancy_gui()
        window.mainloop()

if __name__ == "__main__":
    chatbot = ModernChatbotUI()
    chatbot.run()
```

```python
%pip install numpy

import nltk
import random
import numpy as np # type: ignore
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
import json
import pickle
import re

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

class AdvancedChatbot:
    def __init__(self):
        # Lemmatizer for word normalization
        self.lemmatizer = WordNetLemmatizer()
        
        # Conversation patterns and intents
        self.intents = {
            'greetings': {
                'patterns': [
                    'hi', 'hello', 'hey', 'greetings', 'wassup', 
                    'good morning', 'good afternoon', 'good evening'
                ],
                'responses': [
                    'Hello! How can I assist you today?', 
                    'Hi there! What can I help you with?', 
                    'Greetings! I\'m ready to chat.'
                ]
            },
            'help': {
                'patterns': [
                    'help', 'assist me', 'what can you do', 
                    'support', 'guidance'
                ],
                'responses': [
                    'I can help you with various tasks. Ask me anything!',
                    'I\'m an AI assistant designed to provide information and support.',
                    'I\'m here to help. What do you need assistance with?'
                ]
            },
            'mood': {
                'patterns': [
                    'how are you', 'how\'s it going', 'what\'s up', 
                    'how do you feel'
                ],
                'responses': [
                    'I\'m doing great, thanks for asking!', 
                    'Feeling awesome and ready to help!', 
                    'Always good and eager to assist you!'
                ]
            },
            'unknown': {
                'patterns': [],
                'responses': [
                    'I\'m not sure I understand. Could you rephrase that?',
                    'That\'s an interesting query. Could you provide more context?',
                    'I\'m still learning. Can you explain that differently?'
                ]
            }
        }
        
        # Neural network model configuration
        self.max_words = 1000
        self.max_len = 20
        
    def preprocess_text(self, text):
        """
        Advanced text preprocessing
        """
        # Convert to lowercase
        text = text.lower()
        
        # Remove special characters
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        
        # Tokenize
        tokens = nltk.word_tokenize(text)
        
        # Lemmatization
        tokens = [self.lemmatizer.lemmatize(token) for token in tokens]
        
        return ' '.join(tokens)
    
    def create_training_data(self):
        """
        Prepare training data for the neural network
        """
        # Collect all patterns
        training_sentences = []
        training_labels = []
        
        for intent, data in self.intents.items():
            for pattern in data['patterns']:
                training_sentences.append(self.preprocess_text(pattern))
                training_labels.append(intent)
        
        # Tokenization
        tokenizer = Tokenizer(num_words=self.max_words, oov_token='<OOV>')
        tokenizer.fit_on_texts(training_sentences)
        
        # Convert text to sequences
        sequences = tokenizer.texts_to_sequences(training_sentences)
        padded_sequences = pad_sequences(sequences, maxlen=self.max_len, padding='post')
        
        # Encode labels
        label_encoder = LabelEncoder()
        encoded_labels = label_encoder.fit_transform(training_labels)
        
        return padded_sequences, encoded_labels, tokenizer, label_encoder
    
    def build_model(self, input_shape, num_classes):
        """
        Create neural network model
        """
        model = keras.Sequential([
            keras.layers.Embedding(self.max_words, 16, input_length=input_shape),
            keras.layers.GlobalAveragePooling1D(),
            keras.layers.Dense(24, activation='relu'),
            keras.layers.Dense(num_classes, activation='softmax')
        ])
        
        model.compile(
            optimizer='adam', 
            loss='sparse_categorical_crossentropy', 
            metrics=['accuracy']
        )
        
        return model
    
    def train_chatbot(self):
        """
        Train the neural network
        """
        # Prepare training data
        X, y, tokenizer, label_encoder = self.create_training_data()
        
        # Build and train model
        model = self.build_model(X.shape[1], len(np.unique(y)))
        model.fit(X, y, epochs=50, verbose=0)
        
        return model, tokenizer, label_encoder
    
    def get_response(self, user_input):
        """
        Generate chatbot response
        """
        # Preprocess input
        processed_input = self.preprocess_text(user_input)
        
        # Tokenize and pad input
        input_sequence = self.tokenizer.texts_to_sequences([processed_input])
        padded_input = pad_sequences(input_sequence, maxlen=self.max_len, padding='post')
        
        # Predict intent
        prediction = self.model.predict(padded_input)
        intent_index = np.argmax(prediction)
        intent = self.label_encoder.inverse_transform([intent_index])[0]
        
        # Select response
        if intent in self.intents:
            return random.choice(self.intents[intent]['responses'])
        else:
            return random.choice(self.intents['unknown']['responses'])
    
    def interactive_chat(self):
        """
        Interactive chat interface
        """
        print("🤖 Advanced AI Chatbot: Type 'bye' to exit")
        
        # Train model
        self.model, self.tokenizer, self.label_encoder = self.train_chatbot()
        
        while True:
            user_input = input("You: ")
            
            if user_input.lower() == 'bye':
                print("Chatbot: Goodbye! Have a great day!")
                break
            
            response = self.get_response(user_input)
            print("Chatbot:", response)

# Run the chatbot
if __name__ == "__main__":
    chatbot = AdvancedChatbot()
    chatbot.interactive_chat()
```
